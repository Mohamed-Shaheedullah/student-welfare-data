```python
# to avoid boolean error vectorize
@np.vectorize  
def replace_non_Likert(x):
    # next line to avoid type errors
    if x is not None and not math.isnan(x):
        if np.logical_and(x>0 , x < 10):
            return x
        return np.NaN
```
